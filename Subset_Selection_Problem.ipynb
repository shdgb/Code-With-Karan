```python
from itertools import combinations

# Set of numbers
# S = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]
#
# # Problem 1: Subsets of size 5 with sum 0
# result_5 = [subset for subset in combinations(S, 5) if sum(subset) == 0]
#
# # Display result
# print(f"Total subsets of size 5 with sum 0: {len(result_5)}")
# for subset in result_5:
#     print(subset)


from itertools import combinations

# Define the input set
S = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]

# Create empty list to store results
result_3_to_6 = []

# Check all subset sizes from 3 to 6
for r in range(3, 7):
    result_3_to_6.extend([subset for subset in combinations(S, r) if sum(subset) == 0])

# Print the results
print(f"Total subsets of size 3 to 6 with sum 0: {len(result_3_to_6)}")
for subset in result_3_to_6:
    print(subset)
```
