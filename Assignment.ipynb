```python
import random
import math
import time
import matplotlib.pyplot as plt

# Q1
L = [11, 12, 13, 14]

# (i)
L.extend([50, 60])
print("Q1 (i):", L)

# (ii)
if 11 in L:
    L.remove(11)
if 13 in L:
    L.remove(13)
print("Q1 (ii):", L)

# (iii)
L.sort()
print("Q1 (iii):", L)

# (iv)
L.sort(reverse=True)
print("Q1 (iv):", L)

# (v)
if 13 in L:
    print("Q1 (v): 13 is present")
else:
    print("Q1 (v): 13 is not present")

# (vi)
print("Q1 (vi):", len(L))

# (vii)
print("Q1 (vii):", sum(L))

# (viii)
odd_sum = sum(x for x in L if x % 2 != 0)
print("Q1 (viii):", odd_sum)

# (ix)
even_sum = sum(x for x in L if x % 2 == 0)
print("Q1 (ix):", even_sum)

# (x)
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

prime_sum = sum(x for x in L if is_prime(x))
print("Q1 (x):", prime_sum)

# (xi)
L.clear()
print("Q1 (xi):", L)

# (xii)
del L
#print("Q1 (xii): L is deleted") # Accessing L after deletion will cause an error.

# Q2
D = {1: 5.6, 2: 7.8, 3: 6.6, 4: 8.7, 5: 7.7}

# (i)
D[8] = 8.8
print("Q2 (i):", D)

# (ii)
if 2 in D:
    del D[2]
print("Q2 (ii):", D)

# (iii)
if 6 in D:
    print("Q2 (iii): 6 is present")
else:
    print("Q2 (iii): 6 is not present")

# (iv)
print("Q2 (iv):", len(D))

# (v)
print("Q2 (v):", sum(D.values()))

# (vi)
D[3] = 7.1
print("Q2 (vi):", D)

# (vii)
D.clear()
print("Q2 (vii):", D)

# Q3
S1 = {10, 20, 30, 40, 50, 60}
S2 = {40, 50, 60, 70, 80, 90}

# (i)
S1.update([55, 66])
print("Q3 (i):", S1)

# (ii)
S1.discard(10)
S1.discard(30)
print("Q3 (ii):", S1)

# (iii)
if 40 in S1:
    print("Q3 (iii): 40 is present")
else:
    print("Q3 (iii): 40 is not present")

# (iv)
print("Q3 (iv):", S1.union(S2))

# (v)
print("Q3 (v):", S1.intersection(S2))

# (vi)
print("Q3 (vi):", S1 - S2)

# Q4
# (i)
print("Q4 (i):")
for _ in range(100):
    length = random.randint(6, 8)
    random_string = ''.join(random.choice('abcdefghijklmnopqrstuvwxyz') for _ in range(length))
    print(random_string)

# (ii)
print("Q4 (ii):")
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

for num in range(600, 801):
    if is_prime(num):
        print(num)

# (iii)
print("Q4 (iii):")
for num in range(100, 1001):
    if num % 7 == 0 and num % 9 == 0:
        print(num)

# Q5
list1 = [random.randint(10, 30) for _ in range(10)]
list2 = [random.randint(10, 30) for _ in range(10)]

print("Q5: List 1:", list1)
print("Q5: List 2:", list2)

# (i)
common_numbers = list(set(list1) & set(list2))
print("Q5 (i): Common numbers:", common_numbers)

# (ii)
unique_numbers = list(set(list1) ^ set(list2))
print("Q5 (ii): Unique numbers:", unique_numbers)

# (iii)
print("Q5 (iii): Minimum in List 1:", min(list1))
print("Q5 (iii): Minimum in List 2:", min(list2))

# (iv)
print("Q5 (iv): Maximum in List 1:", max(list1))
print("Q5 (iv): Maximum in List 2:", max(list2))

# (v)
print("Q5 (v): Sum of List 1:", sum(list1))
print("Q5 (v): Sum of List 2:", sum(list2))

# Q6
random_numbers = [random.randint(100, 900) for _ in range(100)]
print("Q6: Random Numbers:", random_numbers)

odd_numbers = [num for num in random_numbers if num % 2 != 0]
even_numbers = [num for num in random_numbers if num % 2 == 0]
prime_numbers = [num for num in random_numbers if is_prime(num)]

print("Q6 (i): Odd numbers count:", len(odd_numbers))
print("Q6 (ii): Even numbers count:", len(even_numbers))
print("Q6 (iii): Prime numbers count:", len(prime_numbers))

# Q7
D = {1: "One", 2: "Two", 3: "Three", 4: "Four", 5: "Five"}

with open("Q7_output.txt", "w") as f:
    for key, value in D.items():
        f.write(f"Key{key}, {value}\n")

print("Q7: Output written to Q7_output.txt")

# Q8
L = ["One", "Two", "Three", "Four", "Five"]

with open("Q8_output.txt", "w") as f:
    for element in L:
        f.write(f"{element}, {len(element)}\n")

print("Q8: Output written to Q8_output.txt")

# Q9
with open("Q9_output.txt", "w") as f:
    for _ in range(100):
        length = random.randint(10, 15)
        random_string = ''.join(random.choice('abcdefghijklmnopqrstuvwxyz') for _ in range(length))
        f.write(f"{random_string}\n")

print("Q9: Output written to Q9_output.txt")

# Q10
with open("Q10_output.txt", "w") as f:
    for num in range(600, 801):
        if is_prime(num):
            f.write(f"{num}\n")

print("Q10: Output written to Q10_output.txt")

# Q11
start_time = time.time()
# Your program code here
time.sleep(2) # Simulate some work
end_time = time.time()
time_taken = end_time - start_time
print("Q11: Time taken:", time_taken, "seconds")

# Q12
def sort_and_time(n):
    list_to_sort = [random.randint(1, 1000) for _ in range(n)]
    start_time = time.time()
    list_to_sort.sort()
    end_time = time.time()
    return end_time - start_time

element_counts = [5000, 10000, 15000, 20000, 25000]
times = []

for count in element_counts:
    time_taken = sort_and_time(count)
    times.append(time_taken)
    print(f"Q12: Time taken to sort {count} elements: {time_taken} seconds")

plt.plot(element_counts, times)
plt.xlabel("Number of elements")
plt.ylabel("Time taken (seconds)")
plt.title("Sorting Time vs. Number of Elements")
plt.show()

# Q13
def find_max_min_average(student_marks):
    """
    Finds the student with the maximum and minimum average marks.

    Args:
        student_marks (dict): A dictionary where keys are student names and values
                              are dictionaries of subject marks.

    Returns:
        tuple: A tuple containing the student with the maximum average marks and
               the student with the minimum average marks.  Returns (None, None)
               if the input dictionary is empty.
    """

    if not student_marks:
        return None, None

    student_averages = {}
    for student, marks in student_marks.items():
        total_marks = sum(marks.values())
        average_marks = total_marks / len(marks) if marks else 0  # Avoid division by zero
        student_averages[student] = average_marks

    max_avg_student = max(student_averages, key=student_averages.get)
    min_avg_student = min(student_averages, key=student_averages.get)

    return max_avg_student, min_avg_student


# Example usage:
student_data = {
    "Alice": {"Math": 85, "Science": 92, "English": 78, "History": 88, "Computer Science": 95},
    "Bob": {"Math": 70, "Science": 80, "English": 90, "History": 75, "Computer Science": 85},
    "Charlie": {"Math": 90, "Science": 85, "English": 80, "History": 92, "Computer Science": 88},
    "David": {"Math": 60, "Science": 70, "English": 65, "History": 70, "Computer Science": 75},
}

max_avg, min_avg = find_max_min_average(student_data)

if max_avg and min_avg:
    print(f"Q13: Student with maximum average marks: {max_avg}")
    print(f"Q13: Student with minimum average marks: {min_avg}")
else:
    print("Q13: No student data available.")
```
